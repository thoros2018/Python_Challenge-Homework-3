## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique = purchases["Item ID"].max()
average = purchases["Price"].mean()
purchase = purchases["Purchase ID"].count()
revenue = purchases["Price"].sum()
```

```python
# Place all of the data found into a summary DataFrame
summary_table = pd.DataFrame({"Number of Unique Items": unique,
                              "Average Price": [average],
                              "Number of Purchases": [purchase],
                              "Total Revenue": revenue})
summary_table
```

```python
summary_table["Average Price"] = summary_table["Average Price"].map('${:,.2f}'.format)
summary_table["Total Revenue"] = summary_table["Total Revenue"].map('${:,.2f}'.format)
```
